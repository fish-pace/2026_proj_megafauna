```python
import earthaccess
import xarray as xr
import pandas as pd
import numpy as np
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt

# =========================
# 1. Read observations
# =========================
obs_file = '/home/jovyan/2026_proj_megafauna/data/OBIS_NEFSC_offshore_obs.csv'
obs = pd.read_csv(obs_file)
obs['date'] = pd.to_datetime(obs['date_time']).dt.date
obs['date'] = pd.to_datetime(obs['date'])

print(f"Loaded {len(obs)} observations, date range: {obs['date'].min()} to {obs['date'].max()}")
unique_dates = obs['date'].unique()
print(f"{len(unique_dates)} unique dates")

# =========================
# 2. Login to Earthdata
# =========================
earthaccess.login(persist=True)

# =========================
# 3. Search weekly SSH anomaly files
# =========================
print(f"\nSearching weekly SSH anomaly files from {obs['date'].min()} to {obs['date'].max()}...")
results = earthaccess.search_data(
    short_name="NASA_SSH_REF_SIMPLE_GRID_V1",
    temporal=(str(obs['date'].min().date()), str(obs['date'].max().date()))
)
print(f"Found {len(results)} weekly SSH files")

# Download the data
fileset = earthaccess.open(results)

# =========================
# 4. Match observations to nearest weekly SSH grid cell
# =========================
ssh_records = []

for file in tqdm(fileset, desc="Processing SSH files"):
    with xr.open_dataset(file) as ds:
        # Each file has a single weekly time
        time_week = pd.to_datetime(ds.time.item())
        lat_vals = ds.latitude.values
        lon_vals = ds.longitude.values

        # Find observations within this week
        week_start = time_week
        week_end = time_week + pd.Timedelta(days=7)
        obs_this_week = obs[(obs['date'] >= week_start) & (obs['date'] < week_end)]

        for idx, row in obs_this_week.iterrows():
            lat_idx = np.abs(lat_vals - row['latitude']).argmin()
            lon_idx = np.abs(lon_vals - row['longitude']).argmin()

            ssh_value = float(ds.ssha.isel(latitude=lat_idx, longitude=lon_idx).values)

            ssh_records.append({
                'Original_Index': idx,
                'date_time': row['date_time'],
                'date': row['date'],
                'common_name': row['common_name'],
                'scientific_name': row['scientific_name'],
                'group_size': row['group_size'],
                'latitude': row['latitude'],
                'longitude': row['longitude'],
                'ssha': ssh_value
            })

# =========================
# 5. Save results CSV
# =========================
if ssh_records:
    ssh_df = pd.DataFrame(ssh_records)
    output_file = "/home/jovyan/2026_proj_megafauna/data/NEFSC_offshore_obs_with_SSHA.csv"
    ssh_df.to_csv(output_file, index=False)
    print(f"\nSaved CSV with SSH anomaly: {output_file}")
else:
    print("No SSH data matched!")

# =========================
# 6. Plotting
# =========================
if ssh_records:
    ssh_df['date'] = pd.to_datetime(ssh_df['date'])
    sns.set_style("whitegrid")

    # ===== Boxplot =====
    fig, ax = plt.subplots(figsize=(12, 6))
    species_order = ssh_df['common_name'].value_counts().index.tolist()
    bp = ax.boxplot(
        [ssh_df[ssh_df['common_name']==s]['ssha'].dropna() for s in species_order],
        labels=species_order,
        patch_artist=True,
        showmeans=True,
        meanprops=dict(marker='D', markerfacecolor='red', markersize=8)
    )
    colors = ['lightblue', 'lightgreen', 'lightcoral'] * ((len(species_order)//3)+1)
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)
        patch.set_alpha(0.7)
    ax.set_xlabel('Species', fontsize=12, fontweight='bold')
    ax.set_ylabel('SSH Anomaly (m)', fontsize=12, fontweight='bold')
    ax.set_title('SSH Anomaly Distribution by Species', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    ax.grid(True, alpha=0.3, axis='y')
    plt.tight_layout()
    plt.savefig('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/SSH_boxplot.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ===== Violin Plot =====
    fig, ax = plt.subplots(figsize=(12, 6))
    sns.violinplot(data=ssh_df, x='common_name', y='ssha', palette='Set2', ax=ax)
    sns.stripplot(data=ssh_df, x='common_name', y='ssha', color='black', alpha=0.3, size=3, ax=ax)
    ax.set_xlabel('Species', fontsize=12, fontweight='bold')
    ax.set_ylabel('SSH Anomaly (m)', fontsize=12, fontweight='bold')
    ax.set_title('SSH Anomaly Distribution by Species (Violin Plot)', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    ax.grid(True, alpha=0.3, axis='y')
    plt.tight_layout()
    plt.savefig('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/SSH_violinplot.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ===== Summary statistics =====
    summary = ssh_df.groupby('common_name')['ssha'].agg([
        ('Count','count'),
        ('Mean','mean'),
        ('Median','median'),
        ('Std','std'),
        ('Min','min'),
        ('Max','max'),
        ('25%', lambda x: x.quantile(0.25)),
        ('75%', lambda x: x.quantile(0.75))
    ]).round(3)
    print("\n=== SSH Summary by Species ===")
    print(summary)
    summary.to_csv('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/SSH_summary_stats.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# =========================
# 1. Load SSH observation data
# =========================
ssh_file = '/home/jovyan/2026_proj_megafauna/data/NEFSC_offshore_obs_with_SSHA.csv'
ssh_df = pd.read_csv(ssh_file)

# Clean data
analysis_df = ssh_df[['common_name', 'group_size', 'ssha']].dropna()
analysis_df = analysis_df[analysis_df['group_size'] > 0]

print(f"Total observations for SSH analysis: {len(analysis_df)}")
print(f"Species found: {analysis_df['common_name'].unique()}")

# =========================
# 2. Correlation analysis by species
# =========================
print("\n" + "="*70)
print("CORRELATION ANALYSIS: Group Size vs SSH Anomaly")
print("="*70)

correlation_results = []

species_list = analysis_df['common_name'].unique()

for species in species_list:
    species_data = analysis_df[analysis_df['common_name'] == species]
    
    if len(species_data) > 2:
        pearson_r, pearson_p = stats.pearsonr(species_data['group_size'], species_data['ssha'])
        spearman_r, spearman_p = stats.spearmanr(species_data['group_size'], species_data['ssha'])
        
        print(f"\n{species} (n={len(species_data)})")
        print("-" * 50)
        print(f"  Pearson r = {pearson_r:.3f}, p = {pearson_p:.4f}")
        print(f"  Spearman ρ = {spearman_r:.3f}, p = {spearman_p:.4f}")
        
        if spearman_p < 0.05:
            direction = "positive" if spearman_r > 0 else "negative"
            print(f"  → Significant {direction} correlation (p < 0.05)")
        else:
            print(f"  → No significant correlation (p ≥ 0.05)")
        
        correlation_results.append({
            'Species': species,
            'N': len(species_data),
            'Pearson_r': pearson_r,
            'Pearson_p': pearson_p,
            'Spearman_rho': spearman_r,
            'Spearman_p': spearman_p
        })

# Save correlation results
if correlation_results:
    corr_df = pd.DataFrame(correlation_results)
    corr_df.to_csv('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/groupsize_ssha_correlations.csv', index=False)
    print("\nCorrelation results saved to groupsize_ssha_correlations.csv")

# =========================
# 3. Box plot by species
# =========================
fig, ax = plt.subplots(figsize=(10, 7))
species_order = analysis_df['common_name'].value_counts().index.tolist()

bp = ax.boxplot([analysis_df[analysis_df['common_name'] == sp]['ssha'].dropna() 
                 for sp in species_order],
                 labels=species_order,
                 patch_artist=True,
                 showmeans=True,
                 meanprops=dict(marker='D', markerfacecolor='red', markersize=8))

colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow']
for patch, color in zip(bp['boxes'], colors[:len(bp['boxes'])]):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_xlabel('Species', fontsize=12, fontweight='bold')
ax.set_ylabel('SSH Anomaly (m)', fontsize=12, fontweight='bold')
ax.set_title('SSH Anomaly Distribution by Species', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/ssha_by_species_boxplot.png', dpi=300, bbox_inches='tight')
plt.show()

# =========================
# 4. Box plot by group size category
# =========================
analysis_df['group_size_category'] = pd.cut(analysis_df['group_size'], bins=[0,1,2,4,100], labels=['1','2','3-4','5+'])

fig, ax = plt.subplots(figsize=(12, 7))
sns.boxplot(data=analysis_df, x='group_size_category', y='ssha', hue='common_name', palette='Set2', ax=ax)

ax.set_xlabel('Group Size Category', fontsize=12, fontweight='bold')
ax.set_ylabel('SSH Anomaly (m)', fontsize=12, fontweight='bold')
ax.set_title('SSH Anomaly by Group Size Category and Species', fontsize=14, fontweight='bold')
ax.legend(title='Species', fontsize=10, bbox_to_anchor=(1.05,1), loc='upper left')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/groupsize_categories_ssha.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nAll SSH correlation plots saved successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import kruskal, gaussian_kde

# Load your data
df = pd.read_csv('/home/jovyan/2026_proj_megafauna/data/NEFSC_Offshore_Obs_With_SST_SSHA.csv')

# Clean data for SST and SSHA
df_clean = df[['common_name', 'latitude', 'longitude', 'sst_median', 'sst_std', 'ssha']].dropna()

species_list = df_clean['common_name'].unique()
colors = sns.color_palette("Set2", len(species_list))

print("="*70)
print("ENVIRONMENTAL NICHE ANALYSIS: SST & SSHA")
print("="*70)

# =========================
# Summary statistics
# =========================
summary = df_clean.groupby('common_name').agg({
    'sst_median': ['count', 'mean', 'std', 'min', 'max'],
    'sst_std': ['mean', 'std'],
    'ssha': ['mean', 'std', 'min', 'max'],
    'latitude': ['mean', 'std'],
    'longitude': ['mean', 'std']
}).round(3)

print("\nEnvironmental Summary by Species:")
print(summary)

summary.to_csv('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/Species_Env_Summary_SST_SSHA.csv')

# =========================
# Multi-panel plots
# =========================
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Panel 1: SST Distribution
ax1 = axes[0,0]
for sp, color in zip(species_list, colors):
    data = df_clean[df_clean['common_name'] == sp]['sst_median']
    ax1.hist(data, alpha=0.6, bins=15, color=color, label=sp, edgecolor='black')
ax1.set_title("SST Distribution", fontweight='bold')
ax1.set_xlabel("SST (°C)", fontweight='bold')
ax1.set_ylabel("Frequency", fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Panel 2: SSH Anomaly Distribution
ax2 = axes[0,1]
for sp, color in zip(species_list, colors):
    data = df_clean[df_clean['common_name'] == sp]['ssha']
    ax2.hist(data, alpha=0.6, bins=15, color=color, label=sp, edgecolor='black')
ax2.set_title("SSH Anomaly (SSHA) Distribution", fontweight='bold')
ax2.set_xlabel("SSHA (m)", fontweight='bold')
ax2.set_ylabel("Frequency", fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.legend()

# Panel 3: Latitude Distribution
ax3 = axes[1,0]
for sp, color in zip(species_list, colors):
    data = df_clean[df_clean['common_name'] == sp]['latitude']
    ax3.hist(data, alpha=0.6, bins=15, color=color, label=sp, edgecolor='black')
ax3.set_title("Latitude Distribution", fontweight='bold')
ax3.set_xlabel("Latitude (°N)", fontweight='bold')
ax3.set_ylabel("Frequency", fontweight='bold')
ax3.grid(True, alpha=0.3)
ax3.legend()

# Panel 4: SST vs SSHA (Environmental Space)
ax4 = axes[1,1]
for sp, color in zip(species_list, colors):
    sp_data = df_clean[df_clean['common_name'] == sp]
    ax4.scatter(sp_data['sst_median'], sp_data['ssha'], 
                alpha=0.6, s=50, label=sp, color=color, edgecolors='black', linewidths=0.5)
ax4.set_title("Environmental Space: SST vs SSHA", fontweight='bold')
ax4.set_xlabel("SST (°C)", fontweight='bold')
ax4.set_ylabel("SSHA (m)", fontweight='bold')
ax4.grid(True, alpha=0.3)
ax4.legend()

plt.tight_layout()
plt.savefig('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/Env_Niche_SST_SSHA.png', dpi=300, bbox_inches='tight')
plt.show()

# =========================
# Kruskal-Wallis tests
# =========================
print("\nSTATISTICAL TESTS: Kruskal-Wallis")

# SST
groups_sst = [df_clean[df_clean['common_name']==sp]['sst_median'].values for sp in species_list]
h_sst, p_sst = kruskal(*groups_sst)
print(f"SST Concentration: H={h_sst:.3f}, p={p_sst:.4f} -> {'Significant' if p_sst<0.05 else 'Not significant'}")

# SSHA
groups_ssha = [df_clean[df_clean['common_name']==sp]['ssha'].values for sp in species_list]
h_ssha, p_ssha = kruskal(*groups_ssha)
print(f"SSHA: H={h_ssha:.3f}, p={p_ssha:.4f} -> {'Significant' if p_ssha<0.05 else 'Not significant'}")

# =========================
# Niche overlap (Pianka's index)
# =========================
def calculate_overlap(data1, data2):
    """Kernel density based niche overlap"""
    x_min = min(data1.min(), data2.min())
    x_max = max(data1.max(), data2.max())
    x = np.linspace(x_min, x_max, 1000)
    kde1 = gaussian_kde(data1)
    kde2 = gaussian_kde(data2)
    density1 = kde1(x)
    density2 = kde2(x)
    overlap = np.sum(np.minimum(density1, density2)) / np.sqrt(np.sum(density1**2) * np.sum(density2**2))
    return overlap

print("\nNICHE OVERLAP (SST):")
for i, sp1 in enumerate(species_list):
    for sp2 in species_list[i+1:]:
        d1 = df_clean[df_clean['common_name']==sp1]['sst_median'].values
        d2 = df_clean[df_clean['common_name']==sp2]['sst_median'].values
        print(f"{sp1} vs {sp2}: {calculate_overlap(d1,d2):.3f}")

print("\nNICHE OVERLAP (SSHA):")
for i, sp1 in enumerate(species_list):
    for sp2 in species_list[i+1:]:
        d1 = df_clean[df_clean['common_name']==sp1]['ssha'].values
        d2 = df_clean[df_clean['common_name']==sp2]['ssha'].values
        print(f"{sp1} vs {sp2}: {calculate_overlap(d1,d2):.3f}")

print("\nAnalysis complete! Summary saved and plots generated.")
```
