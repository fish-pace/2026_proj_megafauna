```python
import earthaccess
import xarray as xr
import pandas as pd

# 1. Login
earthaccess.login(persist=True)

# 2. Search weekly SSH reference grid
results = earthaccess.search_data(
    short_name="NASA_SSH_REF_SIMPLE_GRID_V1",
    temporal=("2024-12-01", "2025-03-30")
)

print(f"Weekly granules found: {len(results)}")

# 3. Download
files = earthaccess.download(results)

# 4. Build time vector from granule metadata (ORDER MATTERS)
times = [
    pd.to_datetime(g["umm"]["TemporalExtent"]["RangeDateTime"]["BeginningDateTime"])
    for g in results
]

# 5. Preprocess function to attach time BEFORE concat
def add_time(ds, time):
    ds = ds.expand_dims(time=[time])
    return ds

# 6. Open datasets cleanly
ds = xr.open_mfdataset(
    files,
    combine="nested",
    concat_dim="time",
    preprocess=lambda d, t=iter(times): add_time(d, next(t))
)

# 7. Fix longitude (0‚Äì360 ‚Üí -180‚Äì180)
if ds.longitude.max() > 180:
    ds = ds.assign_coords(
        longitude=((ds.longitude + 180) % 360) - 180
    ).sortby("longitude")

# 8. Subset bounding box
ds = ds.sel(
    longitude=slice(-77, -71.5),
    latitude=slice(35, 41)
)

# 9. Select SSH anomaly
ssh = ds["ssha"]

ssh
```

```python
import matplotlib.pyplot as plt

# SSH anomaly (ensure numeric)
ssh = ds['ssha'].astype('float32')

# Spatial subset
ssh_sub = ssh.sel(
    latitude=slice(35.0, 41.0),
    longitude=slice(-77.0, -71.5)
)

# ---- Average ALL weeks together ----
ssh_weekly_mean = ssh_sub.mean(dim='time', skipna=True)

# Subset observations to domain
obs_sub = obs[
    (obs['longitude'] >= -77.0) & (obs['longitude'] <= -71.5) &
    (obs['latitude']  >= 35.0) & (obs['latitude']  <= 41.0)
]

# Plot
fig, ax = plt.subplots(figsize=(8, 6))
ssh_weekly_mean.plot(
    ax=ax,
    cmap='RdBu_r',
    robust=True,
    cbar_kwargs={'label': 'Mean Weekly SSH Anomaly (m)'}
)

# Overlay observations
ax.scatter(
    obs_sub['longitude'],
    obs_sub['latitude'],
    c='black',
    s=50,
    alpha=0.8,
    edgecolors='white',
    linewidths=0.6,
    label='Marine Mammal Observations'
)

ax.set_title('Mean Weekly SSH Anomaly (Study Period)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import xarray as xr

# ======================================================
# Match observations to SSH anomaly timeseries
# ======================================================

# Ensure obs['date'] is datetime
obs['date'] = pd.to_datetime(obs['date'])

# Convert SSH times to datetime64
ssh_times = pd.to_datetime(ssh_sub['time'].values)

results_list = []

print(f"Matching {len(obs)} observations to SSH timeseries...")

for idx, row in tqdm(obs.iterrows(), total=len(obs), desc="Matching SSH by date"):
    obs_date = row['date']

    # Find the nearest time in SSH
    time_idx = np.abs(ssh_times - obs_date).argmin()

    # Interpolate SSH anomaly at the observation location
    ssh_at_point = ssh_sub.isel(time=time_idx).interp(
        latitude=row['latitude'],
        longitude=row['longitude'],
        method="linear"
    ).values.item()

    results_list.append({
        'Original_Index': idx,
        'date': obs_date,
        'common_name': row['common_name'],
        'scientific_name': row['scientific_name'],
        'group_size': row.get('group_size', np.nan),
        'latitude': row['latitude'],
        'longitude': row['longitude'],
        'ssh_anomaly': ssh_at_point
    })

# Convert to DataFrame
final_df = pd.DataFrame(results_list)

# ======================================================
# Aggregate by species and date for plotting & stats
# ======================================================
final_df_agg = (
    final_df
    .dropna(subset=["ssh_anomaly"])
    .groupby(["common_name", "date"])
    .agg(
        ssha_median=("ssh_anomaly", "median"),
        ssha_mean=("ssh_anomaly", "mean"),
        ssha_std=("ssh_anomaly", "std"),
        n_obs=("ssh_anomaly", "count")
    )
    .reset_index()
)

print(f"Aggregated dataframe shape: {final_df_agg.shape}")
final_df_agg.head()

final_df.to_csv(
    '/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/SSH_matched_obs.csv',
    index=False
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# ============================
# Load the SSH-matched dataset
# ============================
final_df = pd.read_csv(
    '/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/SSH_matched_obs.csv'
)

# Filter out NaNs and group_size = 0
analysis_df = final_df[['common_name', 'group_size', 'ssh_anomaly']].dropna()
analysis_df = analysis_df[analysis_df['group_size'] > 0]

print(f"Total SSH observations for analysis: {len(analysis_df)}")
print(f"Species found: {analysis_df['common_name'].unique()}")

# ============================
# CORRELATION ANALYSIS
# ============================
print("\n" + "="*70)
print("CORRELATION ANALYSIS: Group Size vs SSH Anomaly")
print("="*70)

correlation_results = []

for species in analysis_df['common_name'].unique():
    species_data = analysis_df[analysis_df['common_name'] == species]
    
    if len(species_data) > 2:
        # Pearson correlation
        pearson_r, pearson_p = stats.pearsonr(species_data['group_size'], 
                                              species_data['ssh_anomaly'])
        # Spearman correlation
        spearman_r, spearman_p = stats.spearmanr(species_data['group_size'], 
                                                 species_data['ssh_anomaly'])
        
        print(f"\n{species} (n={len(species_data)})")
        print("-"*50)
        print(f"  Pearson r = {pearson_r:.3f}, p = {pearson_p:.4f}")
        print(f"  Spearman rho = {spearman_r:.3f}, p = {spearman_p:.4f}")
        
        if spearman_p < 0.05:
            direction = "positive" if spearman_r > 0 else "negative"
            print(f"  ‚Üí Significant {direction} correlation (p < 0.05)")
        else:
            print(f"  ‚Üí No significant correlation (p ‚â• 0.05)")
        
        correlation_results.append({
            'Species': species,
            'N': len(species_data),
            'Pearson_r': pearson_r,
            'Pearson_p': pearson_p,
            'Spearman_rho': spearman_r,
            'Spearman_p': spearman_p
        })

# Save correlation results
if correlation_results:
    corr_df = pd.DataFrame(correlation_results)
    corr_df.to_csv(
        '/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/groupsize_ssh_correlations.csv',
        index=False
    )
    print("\nCorrelation results saved to groupsize_ssh_correlations.csv")

# ============================
# BOX PLOT BY SPECIES
# ============================
fig, ax = plt.subplots(figsize=(10, 7))
species_order = analysis_df['common_name'].value_counts().index.tolist()

bp = ax.boxplot(
    [analysis_df[analysis_df['common_name'] == species]['ssh_anomaly'].dropna() 
     for species in species_order],
    labels=species_order,
    patch_artist=True,
    showmeans=True,
    meanprops=dict(marker='D', markerfacecolor='red', markersize=8)
)

colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow']
for patch, color in zip(bp['boxes'], colors[:len(bp['boxes'])]):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_xlabel('Species', fontsize=12, fontweight='bold')
ax.set_ylabel('SSH Anomaly (m)', fontsize=12, fontweight='bold')
ax.set_title(
    'SSH Anomaly Distribution by Species\n(Box: IQR, Red Diamond: Mean, Line: Median)', 
    fontsize=14, fontweight='bold'
)
plt.xticks(rotation=45, ha='right')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig(
    '/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/ssh_by_species_boxplot.png',
    dpi=300, bbox_inches='tight'
)
plt.show()

# ============================
# BOX PLOT BY GROUP SIZE CATEGORY AND SPECIES
# ============================
analysis_df['group_size_category'] = pd.cut(
    analysis_df['group_size'],
    bins=[0, 1, 2, 4, 100],
    labels=['1', '2', '3-4', '5+']
)

fig, ax = plt.subplots(figsize=(12, 7))

sns.boxplot(
    data=analysis_df,
    x='group_size_category',
    y='ssh_anomaly',
    hue='common_name',
    palette='Set2',
    ax=ax
)

ax.set_xlabel('Group Size Category', fontsize=12, fontweight='bold')
ax.set_ylabel('SSH Anomaly (m)', fontsize=12, fontweight='bold')
ax.set_title('SSH Anomaly by Group Size Category and Species', fontsize=14, fontweight='bold')
ax.legend(title='Species', fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig(
    '/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/groupsize_categories_ssh.png',
    dpi=300, bbox_inches='tight'
)
plt.show()

# ============================
# VIOLIN PLOT BY SPECIES
# ============================
fig, ax = plt.subplots(figsize=(10, 7))

sns.violinplot(
    data=analysis_df,
    x='common_name',
    y='ssh_anomaly',
    palette='Set2',
    ax=ax
)

sns.stripplot(
    data=analysis_df,
    x='common_name',
    y='ssh_anomaly',
    color='black',
    alpha=0.3,
    size=3,
    ax=ax
)

ax.set_xlabel('Species', fontsize=12, fontweight='bold')
ax.set_ylabel('SSH Anomaly (m)', fontsize=12, fontweight='bold')
ax.set_title('SSH Anomaly Distribution by Species (Violin Plot)', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig(
    '/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/ssh_violinplot.png',
    dpi=300, bbox_inches='tight'
)
plt.show()

print("\nAll SSH plots saved successfully!")

```

## 3.Environmental Niche Comparison
The Environmental Niche Comparison quantifies whether different marine mammal species occupy distinct ecological niches by comparing the environmental conditions (sst, sst variability, and geographic location) at observation sites. The analysis produces distribution histograms showing the frequency of observations across environmental gradients for each species, statistical tests (Kruskal-Wallis) to determine if species differ significantly in their habitat use, and niche overlap metrics (Pianka's index, 0-1 scale) that quantify how much species' environmental preferences overlap. High niche overlap (>0.7) suggests species are ecological generalists using similar habitats and may compete for resources, while low overlap (<0.4) indicates niche partitioning where species specialize in different conditions, reducing competition. The environmental space plots reveal whether species cluster in different regions of sst-latitude space, indicating habitat specialization, or if they overlap broadly, suggesting opportunistic habitat use. This analysis directly addresses whether marine mammal species in your study area are responding to environmental gradients in species-specific ways (requiring separate conservation strategies) or if they function as a generalized "marine mammal community" with similar habitat requirements (allowing for unified management approaches). The results provide the ecological justification for whether subsequent analyses should be species-specific or can be conducted on pooled data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# Load your SSH-matched dataset
df = pd.read_csv('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/SSH_matched_obs.csv')

# Clean data: keep only species, location, and SSH anomaly
df_clean = df[['common_name', 'latitude', 'longitude', 'ssh_anomaly']].dropna()

print("="*70)
print("SSH ENVIRONMENTAL NICHE ANALYSIS")
print("="*70)

# Summary statistics by species
print("\nSSH Anomaly Conditions by Species:")
summary = df_clean.groupby('common_name').agg({
    'ssh_anomaly': ['count', 'mean', 'std', 'min', 'max'],
    'latitude': ['mean', 'std'],
    'longitude': ['mean', 'std']
}).round(3)

print(summary)

# ===== VISUALIZATION 1: Multi-panel Environmental Comparison =====
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

species_list = df_clean['common_name'].unique()
colors = sns.color_palette("Set2", len(species_list))

# Panel 1: SSH Anomaly Distribution
ax1 = axes[0]
for species, color in zip(species_list, colors):
    data = df_clean[df_clean['common_name'] == species]['ssh_anomaly']
    ax1.hist(data, alpha=0.6, label=species, bins=15, color=color, edgecolor='black')
ax1.set_xlabel('SSH Anomaly (m)', fontweight='bold')
ax1.set_ylabel('Frequency', fontweight='bold')
ax1.set_title('SSH Anomaly Distribution', fontweight='bold', fontsize=13)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Panel 2: Latitude Distribution
ax2 = axes[1]
for species, color in zip(species_list, colors):
    data = df_clean[df_clean['common_name'] == species]['latitude']
    ax2.hist(data, alpha=0.6, label=species, bins=15, color=color, edgecolor='black')
ax2.set_xlabel('Latitude (¬∞N)', fontweight='bold')
ax2.set_ylabel('Frequency', fontweight='bold')
ax2.set_title('Geographic Distribution (North-South)', fontweight='bold', fontsize=13)
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/SSH_Niche_Comparison.png',
            dpi=300, bbox_inches='tight')
plt.show()

# ===== STATISTICAL TESTS: Do species differ in habitat use? =====
print("\n" + "="*70)
print("STATISTICAL COMPARISON: Do species use different SSH habitats?")
print("="*70)

from scipy.stats import kruskal

# Test for SSH anomaly
groups_ssh = [df_clean[df_clean['common_name'] == sp]['ssh_anomaly'].values 
              for sp in species_list]
h_stat_ssh, p_val_ssh = kruskal(*groups_ssh)

print(f"\nSSH Anomaly:")
print(f"  H-statistic: {h_stat_ssh:.3f}")
print(f"  p-value: {p_val_ssh:.4f}")
if p_val_ssh < 0.05:
    print("  ‚Üí Species differ significantly in SSH anomaly use (p < 0.05)")
else:
    print("  ‚Üí No significant difference between species (p ‚â• 0.05)")

# ===== NICHE OVERLAP CALCULATION =====
print("\n" + "="*70)
print("NICHE OVERLAP ANALYSIS (SSH)")
print("="*70)

from scipy.stats import gaussian_kde

def calculate_overlap(species1_data, species2_data):
    """Calculate niche overlap using kernel density estimation"""
    x_min = min(species1_data.min(), species2_data.min())
    x_max = max(species1_data.max(), species2_data.max())
    x = np.linspace(x_min, x_max, 1000)
    
    kde1 = gaussian_kde(species1_data)
    kde2 = gaussian_kde(species2_data)
    
    density1 = kde1(x)
    density2 = kde2(x)
    
    overlap = np.sum(np.minimum(density1, density2)) / np.sqrt(
        np.sum(density1**2) * np.sum(density2**2))
    
    return overlap

if len(species_list) >= 2:
    print("\nSSH Niche Overlap (0 = no overlap, 1 = complete overlap):")
    for i, sp1 in enumerate(species_list):
        for sp2 in species_list[i+1:]:
            data1 = df_clean[df_clean['common_name'] == sp1]['ssh_anomaly'].values
            data2 = df_clean[df_clean['common_name'] == sp2]['ssh_anomaly'].values
            overlap = calculate_overlap(data1, data2)
            print(f"  {sp1} vs {sp2}: {overlap:.3f}")

# Save summary statistics
summary.to_csv('/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/SSH_Species_Summary.csv')

print("\n" + "="*70)
print("SSH NICHE ANALYSIS COMPLETE!")
print("="*70)
```

Interpretation of results: The analysis reveals that marine mammal species in your study area differentiate their niches primarily through geographic location rather than SST conditions. While all species occupy a similar range of SST concentrations (mean 0.64-0.83 mg/m¬≥) with no significant differences between them (p = 0.091), they show significant geographic separation by latitude (p = 0.0028), with species like Blue whales and Sperm whales occurring farther south (~37¬∞N) and Beaked whales and North Atlantic right whales farther north (~38.3-38.4¬∞N). Niche overlap values are generally moderate to low (0.15-0.59), with the two most abundant species‚ÄîFin whales and Humpback whales‚Äîshowing moderate overlap (0.508), indicating they use similar but not identical habitats and may partition resources through fine-scale spatial or temporal differences. Beaked whales are the most distinct, with very low overlap with all other species (<0.17), likely reflecting their deep-water, offshore specialization. The key ecological insight is that these species coexist not by specializing in different productivity regimes, but by occupying different geographic zones within the shelf-slope system, suggesting that bathymetric features (depth, distance to shore, shelf break position) rather than surface productivity drive niche partitioning. This supports analyzing species separately in subsequent models, as each appears to respond to location-specific features rather than following a universal relationship with SST.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import kruskal, gaussian_kde

# =========================
# LOAD DATA
# =========================
df = pd.read_csv(
    '/home/jovyan/2026_proj_megafauna/data/NEFSC_Offshore_Obs_With_SST.csv'
)

# =========================
# TIME HANDLING
# =========================
# CHANGE COLUMN NAME IF NEEDED
df['date'] = pd.to_datetime(df['date'])

# Extract components
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

# =========================
# CREATE CONTINUOUS Dec‚ÄìMar TIME AXIS
# =========================
# Dec = 0‚Äì30, Jan = 31‚Äì61, Feb = 62‚Äì90, Mar = 91‚Äì121
month_offsets = {
    12: 0,    # December
    1: 31,    # January
    2: 62,    # February
    3: 91     # March
}

df['seasonal_doy'] = df.apply(
    lambda row: month_offsets.get(row['month'], np.nan) + row['day'] - 1,
    axis=1
)

# =========================
# CLEAN DATA
# =========================
df_clean = df[
    [
        'common_name',
        'sst_median',
        'sst_std',
        'seasonal_doy'
    ]
].dropna()

species_list = df_clean['common_name'].unique()

print("=" * 70)
print("TEMPORAL (Dec‚ÄìMar) √ó SST NICHE ANALYSIS")
print("=" * 70)

# =========================
# SUMMARY STATISTICS
# =========================
summary = df_clean.groupby('common_name').agg(
    n_obs=('sst_median', 'count'),
    sst_mean=('sst_median', 'mean'),
    sst_std=('sst_median', 'std'),
    sst_min=('sst_median', 'min'),
    sst_max=('sst_median', 'max'),
    time_mean=('seasonal_doy', 'mean'),
    time_std=('seasonal_doy', 'std')
).round(3)

print("\nSummary by species:")
print(summary)

# =========================
# VISUALIZATION
# =========================
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
colors = sns.color_palette("Set2", len(species_list))

# -------------------------
# PANEL 1: DOY √ó SST Scatter
# -------------------------
ax1 = axes[0]
for sp, color in zip(species_list, colors):
    sp_data = df_clean[df_clean['common_name'] == sp]
    ax1.scatter(
        sp_data['seasonal_doy'],
        sp_data['sst_median'],
        s=45,
        alpha=0.65,
        label=sp,
        color=color,
        edgecolors='black',
        linewidths=0.3
    )

ax1.set_xlabel('Seasonal Day (Dec ‚Üí Mar)', fontweight='bold')
ax1.set_ylabel('SST (C)', fontweight='bold')
ax1.set_title('Temporal Environmental Space', fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend()

# -------------------------
# PANEL 2: Temporal Niche Density
# -------------------------
ax2 = axes[1]
for sp, color in zip(species_list, colors):
    data = df_clean[df_clean['common_name'] == sp]['seasonal_doy']
    sns.kdeplot(
        data,
        ax=ax2,
        fill=True,
        alpha=0.5,
        linewidth=2,
        label=sp,
        color=color
    )

ax2.set_xlabel('Seasonal Day (Dec ‚Üí Mar)', fontweight='bold')
ax2.set_ylabel('Density', fontweight='bold')
ax2.set_title('Seasonal Timing Niche', fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.savefig(
    '/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/DOY_SST_Niche.png',
    dpi=300,
    bbox_inches='tight'
)
plt.show()

# =========================
# STATISTICAL TESTS
# =========================
print("\n" + "=" * 70)
print("STATISTICAL COMPARISONS")
print("=" * 70)

# ---- Timing differences ----
groups_time = [
    df_clean[df_clean['common_name'] == sp]['seasonal_doy'].values
    for sp in species_list
]

h_time, p_time = kruskal(*groups_time)

print("\nSeasonal timing (Dec‚ÄìMar):")
print(f"  H-statistic: {h_time:.3f}")
print(f"  p-value: {p_time:.4f}")

# ---- SST differences ----
groups_sst = [
    df_clean[df_clean['common_name'] == sp]['sst_median'].values
    for sp in species_list
]

h_sst, p_sst = kruskal(*groups_sst)

print("\nSST:")
print(f"  H-statistic: {h_sst:.3f}")
print(f"  p-value: {p_sst:.4f}")

# =========================
# TEMPORAL NICHE OVERLAP
# =========================
print("\n" + "=" * 70)
print("TEMPORAL NICHE OVERLAP (Dec‚ÄìMar)")
print("=" * 70)

def niche_overlap(x1, x2):
    x = np.linspace(min(x1.min(), x2.min()), max(x1.max(), x2.max()), 1000)
    kde1 = gaussian_kde(x1)
    kde2 = gaussian_kde(x2)
    d1 = kde1(x)
    d2 = kde2(x)
    return np.sum(np.minimum(d1, d2)) / np.sqrt(np.sum(d1**2) * np.sum(d2**2))

for i, sp1 in enumerate(species_list):
    for sp2 in species_list[i+1:]:
        d1 = df_clean[df_clean['common_name'] == sp1]['seasonal_doy'].values
        d2 = df_clean[df_clean['common_name'] == sp2]['seasonal_doy'].values
        overlap = niche_overlap(d1, d2)
        print(f"{sp1} vs {sp2}: {overlap:.3f}")

# =========================
# SAVE SUMMARY
# =========================
summary.to_csv(
    '/home/jovyan/2026_proj_megafauna/contributor_folders/AnaVaz/Species_DOY_SST_Summary.csv'
)

print("\n" + "=" * 70)
print("ANALYSIS COMPLETE")
print("=" * 70)
```
